```python
import pandas as pd

data = {
    "cars": ["byd", "deepal", "honda", "bmw"],
    "models": ["2001", "2008", "2015", "2024"],
    "passen": ["4", "6", "6", "5"]
}
data
```

```python
# covert the data dictionary to dataframe
df=pd.DataFrame(data)
df
```
